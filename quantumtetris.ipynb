```python
import numpy as np
from block import Block

class QuantumTetris:
    def __init__(self):
        self.current_state = None
        self.blocks = []
        # to be updated with update function
        self.storedblocks = []
        self.time = 0
    

    def check_if_overlap_on_board(self):
        # check if current block overlaps with stored blocks
        overlap = False
        storedpos = np.array(self.storedblocks)[:,0]
        # find blocks with same x
        idx = np.argwhere(storedpos[:,0] == self.blocks[-1].position[0])
        if len(idx) >= 1:
            if np.max(storedpos[idx,1]) <= self.blocks[-1].position[1]:
                overlap = True
        return overlap


    def upcoming_blocks(self):
        # initialize 5 new blocks when called for the first time
        if self.time == 0:
            self.freshblocks = []
            posnew = [] ; orientnew = []
            for i in range(5):
                p = Block()
                pos, orient = p.get_position_orientation()
                posnew.append(pos); orientnew.append(orient)
            self.freshblocks.append(list(zip(posnew, orientnew)))
        else:
            # remove first block which is now being "played" and add new one
            del self.freshblocks[0]
            p = Block()
            pos, orient = p.get_position_orientation()
            self.freshblocks.append(list(zip(pos, orient)))
        return self.freshblocks

    def store(self):
        self.storedblocks.append(list(zip(self.blocks[-1].position, self.blocks[-1].orientation)))
        return self.storedblocks

     # it is the main game logic
    def update(self):
        self.upcoming_blocks() # getting list of upcoming blocks
        
        falling = True
        while (falling == True):

            # checking for overlap to land current block
            if self.check_if_overlap() == True:
                falling = False # current block stops falling
            else:
                self.blocks[-1].position[1] -= -1 # update current block position
            
            self.time.sleep(1) # to repeat this loop every second

        # after block has landed
        if falling == False:
            # create new block
            self.blocks.append(self.freshblocks[0]) # giving next block


    # get board state
    def get_board_state(self):
        return self.blocks
```
